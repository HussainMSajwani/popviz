```python
import pandas as pd
import numpy as np

reduced = pd.read_csv("/home/shussain/popviz/results/PHATE/run2/PHATE_out.csv")
pops = pd.read_csv("/home/shussain/popviz/genomic_data/pops.csv")
relatedness = pd.read_csv("plink.genome", sep="\s+")

def get_ibs(source):
    daddy = source
    ind_daddy = pops.query(f"ID == '{daddy}'").index[0]
    contains_daddy = np.logical_or((relatedness["IID1"] == daddy), (relatedness["IID2"] == daddy))
    relatedness_to_daddy = relatedness[contains_daddy]
    relatedness_to_daddy
    
    other = [ind["IID1"] if ind["IID1"] != daddy else ind["IID2"] for i, ind in relatedness_to_daddy.iterrows()]
    relatedness_to_daddy["ID"] = other
    
    reduced["ID"] = pops["ID"]
    df = relatedness_to_daddy.set_index("ID").join(reduced.set_index("ID"))
    
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    p = plt.scatter(x=df["PHATE1"], y=df["PHATE2"], c=np.log(df["PI_HAT"]), cmap="Blues")
    plt.scatter(reduced.iloc[ind_daddy, 0], reduced.iloc[ind_daddy, 1], c="red")
    plt.colorbar(p)

get_ibs("HGDP01248")
    
```
